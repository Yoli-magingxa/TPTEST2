```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\cash\\Downloads\\Informal_Settlement_Programme_-1898186875815614491.csv")

# Scatter PLot 
shape_area = df['Shape__Area']
shape_length = df['Shape__Length']
plt.figure(figsize=(8,6))
plt.scatter(shape_area,shape_length,alpha=0.5)
plt.title("Shape Area vs Shape Length")
plt.xlabel("Shape Area ")
plt.ylabel("Shape Length")
plt.show()

#Histogram 
plt.figure(figsize=(10,6))
plt.hist(df['EST_HHC'], bins=50, color='skyblue',edgecolor='black')
plt.title('Distribution of Household')
plt.xlabel('Household Count')
plt.ylabel('Frequency')
plt.show()


#Pie chart
category_counts = df['WATER_SUPP'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of WATER SUPP Categories')
plt.axis('equal')
plt.show()

#Advanced plots
#Pairplot
columns = ['EST_HHC', 'Shape__Area', 'Shape__Length']
sns.pairplot(df[columns])
plt.show()

#Violin plot 
plt.figure(figsize=(10,6))
sns.violinplot(x="REGION",y="EST_HHC", data=df)
plt.title("Distribution of EST_HHC across Different Regions")
plt.xlabel("REGION")
plt.ylabel("EST_HHC")
plt.xticks(rotation=45)
plt.show()

summary_stats = df.describe()
print(summary_stats)
```
